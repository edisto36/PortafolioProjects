```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from matplotlib.patches import Rectangle, Circle, FancyBboxPatch
import matplotlib.patches as mpatches

# Set up the style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Create a comprehensive figure with multiple subplots
fig = plt.figure(figsize=(20, 16))
fig.suptitle('Structured vs Unstructured Problems: Comprehensive Comparison',
             fontsize=24, fontweight='bold', y=0.95)

# 1. Problem Characteristics Comparison (Radar Chart)
ax1 = plt.subplot(2, 3, 1, projection='polar')

# Define characteristics and scores
characteristics = ['Clarity', 'Predictability', 'Solution\nCertainty',
                  'Time to\nSolve', 'Resource\nRequirements', 'Repeatability']
structured_scores = [9, 9, 9, 8, 7, 9]
unstructured_scores = [3, 2, 4, 4, 8, 3]

# Number of variables
N = len(characteristics)

# Compute angle for each axis
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]  # Complete the circle

# Add scores for both problem types
structured_scores += structured_scores[:1]
unstructured_scores += unstructured_scores[:1]

# Plot
ax1.plot(angles, structured_scores, 'o-', linewidth=3, label='Structured Problems', color='#2E86AB')
ax1.fill(angles, structured_scores, alpha=0.25, color='#2E86AB')
ax1.plot(angles, unstructured_scores, 'o-', linewidth=3, label='Unstructured Problems', color='#A23B72')
ax1.fill(angles, unstructured_scores, alpha=0.25, color='#A23B72')

# Add labels
ax1.set_xticks(angles[:-1])
ax1.set_xticklabels(characteristics, fontsize=10)
ax1.set_ylim(0, 10)
ax1.set_title('Problem Characteristics\nComparison', fontsize=14, fontweight='bold', pad=20)
ax1.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))
ax1.grid(True)

# 2. Solution Path Visualization
ax2 = plt.subplot(2, 3, 2)

# Structured problem path (linear)
x_structured = np.linspace(0, 10, 100)
y_structured = 5 + 0.1 * np.sin(x_structured)  # Nearly straight line with minor variation

# Unstructured problem path (chaotic)
x_unstructured = np.linspace(0, 10, 100)
y_unstructured = 5 + 2 * np.sin(x_unstructured * 2) + np.random.normal(0, 0.3, 100)

ax2.plot(x_structured, y_structured, linewidth=4, color='#2E86AB', label='Structured Problem Path')
ax2.plot(x_unstructured, y_unstructured, linewidth=3, color='#A23B72', label='Unstructured Problem Path')

# Add start and end points
ax2.scatter([0, 10], [5, 5.1], s=200, c='#2E86AB', marker='o', zorder=5)
ax2.scatter([0, 10], [5, y_unstructured[-1]], s=200, c='#A23B72', marker='o', zorder=5)

# Add annotations
ax2.annotate('START', xy=(0, 5), xytext=(-1, 6), fontsize=12, fontweight='bold',
            arrowprops=dict(arrowstyle='->', color='black', lw=2))
ax2.annotate('GOAL', xy=(10, 5), xytext=(10.5, 6), fontsize=12, fontweight='bold',
            arrowprops=dict(arrowstyle='->', color='black', lw=2))

ax2.set_xlim(-1, 11)
ax2.set_ylim(0, 10)
ax2.set_xlabel('Time/Effort', fontsize=12)
ax2.set_ylabel('Progress', fontsize=12)
ax2.set_title('Solution Path Comparison', fontsize=14, fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)

# 3. Time and Complexity Distribution
ax3 = plt.subplot(2, 3, 3)

# Generate sample data
np.random.seed(42)
structured_times = np.random.normal(2, 0.5, 1000)  # Quick, consistent
unstructured_times = np.random.exponential(5, 1000)  # Variable, often longer

# Create histograms
ax3.hist(structured_times, bins=30, alpha=0.7, label='Structured Problems',
         color='#2E86AB', density=True)
ax3.hist(unstructured_times, bins=30, alpha=0.7, label='Unstructured Problems',
         color='#A23B72', density=True)

ax3.set_xlabel('Time to Solution (arbitrary units)', fontsize=12)
ax3.set_ylabel('Probability Density', fontsize=12)
ax3.set_title('Time Distribution for\nProblem Resolution', fontsize=14, fontweight='bold')
ax3.legend()
ax3.grid(True, alpha=0.3)

# 4. Examples Matrix
ax4 = plt.subplot(2, 3, 4)
ax4.axis('off')

# Create boxes for examples
structured_examples = [
    'Mathematical\nEquations', 'Payroll\nCalculation', 'Inventory\nManagement',
    'Recipe\nFollowing', 'Tax\nComputation', 'Database\nQueries']

unstructured_examples = [
    'Strategic\nPlanning', 'Creative\nDesign', 'Ethical\nDilemmas',
    'Innovation\nProjects', 'Change\nManagement', 'Crisis\nResponse']

# Draw structured examples
for i, example in enumerate(structured_examples):
    x, y = (i % 3) * 0.3 + 0.05, 0.7 - (i // 3) * 0.3
    box = FancyBboxPatch((x, y), 0.25, 0.2, boxstyle="round,pad=0.02",
                         facecolor='#2E86AB', alpha=0.3, edgecolor='#2E86AB')
    ax4.add_patch(box)
    ax4.text(x + 0.125, y + 0.1, example, ha='center', va='center',
             fontsize=10, fontweight='bold')

# Draw unstructured examples
for i, example in enumerate(unstructured_examples):
    x, y = (i % 3) * 0.3 + 0.05, 0.3 - (i // 3) * 0.3
    box = FancyBboxPatch((x, y), 0.25, 0.2, boxstyle="round,pad=0.02",
                         facecolor='#A23B72', alpha=0.3, edgecolor='#A23B72')
    ax4.add_patch(box)
    ax4.text(x + 0.125, y + 0.1, example, ha='center', va='center',
             fontsize=10, fontweight='bold')

# Add labels
ax4.text(0.5, 0.95, 'STRUCTURED PROBLEMS', ha='center', va='center',
         fontsize=14, fontweight='bold', color='#2E86AB')
ax4.text(0.5, 0.55, 'UNSTRUCTURED PROBLEMS', ha='center', va='center',
         fontsize=14, fontweight='bold', color='#A23B72')

ax4.set_xlim(0, 1)
ax4.set_ylim(0, 1)
ax4.set_title('Problem Examples', fontsize=14, fontweight='bold')

# 5. Decision Tree Visualization
ax5 = plt.subplot(2, 3, 5)
ax5.axis('off')

# Create a simple decision tree
def draw_decision_node(ax, x, y, text, color):
    circle = Circle((x, y), 0.08, facecolor=color, edgecolor='black', alpha=0.7)
    ax.add_patch(circle)
    ax.text(x, y, text, ha='center', va='center', fontsize=8, fontweight='bold')

def draw_arrow(ax, x1, y1, x2, y2, text=''):
    ax.annotate('', xy=(x2, y2), xytext=(x1, y1),
                arrowprops=dict(arrowstyle='->', lw=2, color='black'))
    if text:
        mid_x, mid_y = (x1 + x2) / 2, (y1 + y2) / 2
        ax5.text(mid_x, mid_y + 0.05, text, ha='center', va='bottom', fontsize=8)

# Draw decision tree
draw_decision_node(ax5, 0.5, 0.9, 'Problem\nIdentified', '#FFD700')
draw_arrow(ax5, 0.5, 0.82, 0.3, 0.7)
draw_arrow(ax5, 0.5, 0.82, 0.7, 0.7)

ax5.text(0.2, 0.75, 'Clear\nParameters?', ha='center', va='center', fontsize=9)
ax5.text(0.8, 0.75, 'Ambiguous\nGoals?', ha='center', va='center', fontsize=9)

draw_decision_node(ax5, 0.3, 0.6, 'Apply\nAlgorithm', '#2E86AB')
draw_decision_node(ax5, 0.7, 0.6, 'Creative\nApproach', '#A23B72')

draw_arrow(ax5, 0.3, 0.52, 0.3, 0.4)
draw_arrow(ax5, 0.7, 0.52, 0.7, 0.4)

draw_decision_node(ax5, 0.3, 0.3, 'Optimal\nSolution', '#2E86AB')
draw_decision_node(ax5, 0.7, 0.3, 'Iterative\nRefinement', '#A23B72')

ax5.set_xlim(0, 1)
ax5.set_ylim(0, 1)
ax5.set_title('Problem-Solving Decision Tree', fontsize=14, fontweight='bold')

# 6. Success Rate Comparison
ax6 = plt.subplot(2, 3, 6)

categories = ['Solution\nAccuracy', 'Time\nEfficiency', 'Resource\nUtilization',
              'Reproducibility', 'Scalability']
structured_success = [95, 90, 85, 95, 88]
unstructured_success = [70, 60, 75, 45, 65]

x = np.arange(len(categories))
width = 0.35

bars1 = ax6.bar(x - width/2, structured_success, width, label='Structured Problems',
                color='#2E86AB', alpha=0.8)
bars2 = ax6.bar(x + width/2, unstructured_success, width, label='Unstructured Problems',
                color='#A23B72', alpha=0.8)

# Add value labels on bars
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax6.text(bar.get_x() + bar.get_width()/2., height + 1,
                f'{height}%', ha='center', va='bottom', fontsize=10)

ax6.set_ylabel('Success Rate (%)', fontsize=12)
ax6.set_title('Success Metrics Comparison', fontsize=14, fontweight='bold')
ax6.set_xticks(x)
ax6.set_xticklabels(categories, fontsize=10)
ax6.legend()
ax6.grid(True, alpha=0.3, axis='y')
ax6.set_ylim(0, 100)

# Adjust layout and display
plt.tight_layout()
plt.subplots_adjust(top=0.92)
plt.show()

# Print summary statistics
print("\n" + "="*60)
print("STRUCTURED vs UNSTRUCTURED PROBLEMS: KEY INSIGHTS")
print("="*60)
print(f"Structured Problems - Average Solution Time: {np.mean(structured_times):.2f} units")
print(f"Unstructured Problems - Average Solution Time: {np.mean(unstructured_times):.2f} units")
print(f"Structured Problems - Solution Consistency: {np.std(structured_times):.2f} (lower is better)")
print(f"Unstructured Problems - Solution Consistency: {np.std(unstructured_times):.2f} (lower is better)")
print("\nKey Takeaways:")
print("• Structured problems show consistent, predictable solution paths")
print("• Unstructured problems require more time and creative approaches")
print("• Structured problems have higher success rates in most metrics")
print("• Unstructured problems offer more flexibility but less certainty")
```

```python
def calculate_calories_burned(weight_kg, distance_km):
  """
  Calculates the approximate number of calories burned while walking or running.

  Args:
    weight_kg: Weight of the person in kilograms.
    distance_km: Distance covered in kilometers.

  Returns:
    The approximate number of calories burned.
  """
  # Approximate calories burned per kg per km for walking/running
  calories_per_kg_km = 1.0 # This is a general estimate, can vary

  calories_burned = weight_kg * distance_km * calories_per_kg_km
  return calories_burned

# Example usage:
weight = 70 # kg
distance = 5 # km

burned_calories = calculate_calories_burned(weight, distance)
print(f"Approximate calories burned: {burned_calories:.2f}")
```

```python
# This app is translated from Mastering Shinywidgets
# https://mastering-shiny.org/basic-reactivity.html#reactive-expressions-1
from shiny import App, render, ui
from numpy import random

# Functions we import from stats.py
from stats import freqpoly, t_test

app_ui = ui.page_fluid(
    ui.row(
        ui.column(
            4,
            "Distribution 1",
            ui.input_numeric("n1", label="n", value=1000, min=1),
            ui.input_numeric("mean1", label="µ", value=0, step=0.1),
            ui.input_numeric("sd1", label="σ", value=0.5, min=0.1, step=0.1),
        ),
        ui.column(
            4,
            "Distribution 2",
            ui.input_numeric("n2", label="n", value=1000, min=1),
            ui.input_numeric("mean2", label="µ", value=0, step=0.1),
            ui.input_numeric("sd2", label="σ", value=0.5, min=0.1, step=0.1),
        ),
        ui.column(
            4,
            "Frequency polygon",
            ui.input_numeric("binwidth", label="Bin width", value=0.1, step=0.1),
            ui.input_slider("range", label="range", value=[-3, 3], min=-5, max=5),
        ),
    ),
    ui.row(
        ui.column(9, ui.output_plot("hist")),
        ui.column(3, ui.output_text_verbatim("ttest")),
    ),
)


def server(input, output, session):
    @output
    @render.plot
    def hist():
        print(input.range())
        x1 = random.normal(input.mean1(), input.sd1(), input.n1())
        x2 = random.normal(input.mean2(), input.sd2(), input.n2())
        return freqpoly(x1, x2, input.binwidth(), input.range())

    @output
    @render.text
    def ttest():
        x1 = random.normal(0, 1, 100)
        x2 = random.normal(0, 1, 100)
        return t_test(x1, x2)


app = App(app_ui, server)
```

```python
# stats.py
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def freqpoly(x1, x2, binwidth, xrange):
    """
    Create frequency polygon plots for two datasets.

    Args:
        x1 (array-like): First dataset.
        x2 (array-like): Second dataset.
        binwidth (float): The width of the bins for the frequency polygon.
        xrange (tuple): The range of x-values to consider (min, max).

    Returns:
        matplotlib.figure.Figure: The figure containing the frequency polygon plots.
    """
    fig, ax = plt.subplots()

    bins = np.arange(xrange[0], xrange[1] + binwidth, binwidth)

    ax.hist(x1, bins=bins, density=True, histtype='step', label='Distribution 1', linewidth=2)
    ax.hist(x2, bins=bins, density=True, histtype='step', label='Distribution 2', linewidth=2)

    ax.set_xlabel('Value')
    ax.set_ylabel('Density')
    ax.set_title('Frequency Polygon')
    ax.legend()
    ax.grid(True, alpha=0.5)

    return fig


def t_test(x1, x2):
    """
    Perform an independent samples t-test.

    Args:
        x1 (array-like): First dataset.
        x2 (array-like): Second dataset.

    Returns:
        str: A string reporting the t-statistic and p-value.
    """
    t_statistic, p_value = stats.ttest_ind(x1, x2)
    return f"T-test results:\nT-statistic: {t_statistic:.4f}\nP-value: {p_value:.4f}"
```
